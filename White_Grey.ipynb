```python
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
img = cv2.imread('text.jpg', -1)
```

```python
plt.imshow(img)
```

```python
b,g,r = cv2.split(img)
```

```python
img2 = cv2.merge([r,g,b])
```

```python
my_mask= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img3 = img
```

```python
BLACK = 0
WHITE = 255
i = 411
while i >= 0:
    j = 1775
    while j >= 0:
        if my_mask[i][j] == BLACK or my_mask[i][j] == WHITE:
            j = j - 1
            continue
        elif my_mask[i][j] > 182:
            my_mask[i][j] = BLACK
        else:
            my_mask[i][j] = WHITE
        j = j - 1
    i = i - 1
    
print (i)
```

```python
BLACK = 255
WHITE = 0
i = 411
while i >= 0:
    j = 1775
    while j >= 0:
        k = 2
        while k >= 0:
            if my_mask[i][j] == BLACK:
                img3[i][j][k] = BLACK
            else:
                img3[i][j][k] = img2[i][j][k]
            k = k - 1
        j = j - 1
    i = i - 1
    
print (i)
```
