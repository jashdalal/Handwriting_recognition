```python
import matplotlib.pyplot as plt
import cv2
from PIL import Image
import pytesseract
pytesseract.pytesseract.tesseract_cmd = 'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
```

```python
img = cv2.imread('contrast.jpg', 1)
```

```python
plt.imshow(img)
```

```python
img = cv2.GaussianBlur(img,(5,5),0)
```

```python
plt.imshow(img)
```

```python
size_tup = img.shape
```

```python
b,g,r = cv2.split(img)
```

```python
img2 = cv2.merge([r,g,b])
```

```python
my_mask= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img3 = img
```

```python
BLACK = 255
WHITE = 0
i = size_tup[0] - 1
while i >= 0:
    j = size_tup[1] - 1
    while j >= 0:
        if my_mask[i][j] == BLACK or my_mask[i][j] == WHITE:
            j = j - 1
            continue
        elif my_mask[i][j] > 182:
            my_mask[i][j] = BLACK
        else:
            my_mask[i][j] = WHITE
        j = j - 1
    i = i - 1
    
print (i)
```

```python
BLACK = 0
WHITE = 255
i = size_tup[0] - 1
while i >= 0:
    j = size_tup[1] - 1
    while j >= 0:
        k = size_tup[2] - 1
        while k >= 0:
            if my_mask[i][j] == BLACK:
                img3[i][j][k] = BLACK
            else:
                img3[i][j][k] = img2[i][j][k]
            k = k - 1
        j = j - 1
    i = i - 1
    
print (i)
```
